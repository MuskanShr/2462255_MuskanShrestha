Exploratory Data Analysis- QN1

```python
titanic_dataset["Age"]= titanic_dataset["Age"].fillna(titanic_dataset["Age"].median())
```

```python
titanic_dataset["fare_per_year"]=titanic_dataset["Fare"]/titanic_dataset["Age"]
high_fare_age= titanic_dataset.loc[titanic_dataset["fare_per_year"]>5]
high_fare_age_srt= high_fare_age.sort_values(by="fare_per_year", ascending=False)
result=high_fare_age_srt[["Name","fare_per_year"]]
print(result)
```

```python
titanic_dataset["fare_per_class"]=titanic_dataset["Fare"]/titanic_dataset["Pclass"]
adult_males=titanic_dataset.loc[(titanic_dataset["Sex"]=="male")&(titanic_dataset["Age"]>=18)]
adult_males_srt=adult_males.sort_values(by="fare_per_class", ascending=False)
result=adult_males_srt[["Name","Age","fare_per_class"]]
print(result)
```

3.3 Exploratory Data Analysis with Group-by Method

total fare revenue came from each passenger class

```python
total_fare=titanic_dataset["Fare"].sum()
first_class=titanic_dataset.loc[titanic_dataset["Pclass"]==1,"Fare"].sum()
second_class=titanic_dataset.loc[titanic_dataset["Pclass"]==2,"Fare"].sum()
third_class=titanic_dataset.loc[titanic_dataset["Pclass"]==3,"Fare"].sum()
fare_totals=[first_class,second_class,third_class]
fare_revenue=[i/total_fare * 100 for i in fare_totals]
print(fare_revenue)
```

percent of the total number of passengers on the Titanic belonged to each age group

```python
titanic_age=titanic_dataset["Age"].fillna(titanic_dataset["Age"].median())
def age_check(age):
  if age <18:
    return "child"
  elif age < 65:
    return "adult"
  else:
    return "senior"
titanic_dataset["age_group"]=titanic_dataset["Age"].apply(age_check)
total=len(titanic_dataset)
number_of_passengers= titanic_dataset.groupby("age_group").count()
# print(number_of_passengers)
passengers_in_age_group_percent=(number_of_passengers/total)*100
print(passengers_in_age_group_percent)

```
